```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
path=r'C:\Users\user\Desktop\30220\little.png'
```

```python
path
```

```python
Image.open(path)
```

```python
text = open(r'C:\Users\user\Desktop\30220\thelittleprince.txt').read()
```

```python
text
```

```python
lp_mask = np.array(Image.open(path))
```

```python
lp_mask
```

```python
lp_wc=WordCloud(
background_color =  'White',
mask=lp_mask,
max_font_size=50,
contour_width=5,
contour_color='green').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file(r'C:\Users\user\Desktop\30220\little.png')
```
